```python
def making_youtube_table(channel_name,playlist_name_list,i):
    ###ì±„ë„ ì•„ì´ë”” í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ 
    #DataFrameì— í•œ í–‰ì”© ë“¤ì–´ê°ˆ ì¸ìŠ¤í„´ìŠ¤(ë¦¬ìŠ¤íŠ¸ í˜•ì‹)
    videos_col_list=[ 'playlist_id',
                    'youtube_url',
                    'thumb_img',
                    'video_title',
                    'video_views',
                    'video_id',
                    'date']
            
    input_list=[None for i in range(len(videos_col_list))] #videoí…Œì´ë¸”ì— ë„£ì„ ë¹ˆë¦¬ìŠ¤íŠ¸ ìƒì„±
    
    playlist_input_list=[None,None,None,None]
    
    #qëŠ” ì›í•˜ëŠ” ì±„ë„ ì´ë¦„
    search_response=youtube.search().list(
        q=channel_name,
        type='channel', #ì±„ë„ë§Œ
        part='snippet',
        maxResults=1, #ì±„ë„ ì•„ì´ë””ë§Œ ë½‘ì•„ì˜¬ ê²ƒì´ë¯€ë¡œ í•˜ë‚˜ì˜ ê²°ê´ê°’ë§Œ ìš”êµ¬
        ).execute()

    #ê²€ìƒ‰í•œ ì±„ë„ê³ ìœ ì•„ì´ë””
    channel_id = search_response['items'][0]['snippet']['channelId']

    #ì¬ìƒëª©ë¡ì´ ìˆëŠ” ê²½ìš°
    #####################################
    global is_video_list
    if(is_video_list[i] == 1):
        #ì±„ë„ ì•„ì´ë””ë¥¼ ì´ìš©í•˜ì—¬ ì±„ë„ í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ë¡œ ë°›ì•„ì˜´
        playlists = youtube.playlists().list(
            channelId = channel_id,
            part = 'snippet',
            maxResults=50
            ).execute()
        
        #í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ê°€ 2ê°œ ì´ìƒ ìˆì„ ê²½ìš°ë¥¼ ëŒ€ë¹„í•˜ì—¬ ì±„ë„ë‹¹ í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ ì´ë¦„ì„ ë¦¬ìŠ¤íŠ¸í™” í–ˆìŒ
        playlist_id=''
        for playlist_name in playlist_name_list[i]:
            #ì›í•˜ëŠ” í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ ê³ ìœ  ì•„ì´ë”” ì¶”ì¶œ
            for dic in playlists['items']:
                if(dic['snippet']['title']==playlist_name):
                    playlist_id = dic['id']
                    break;
            
            #playlist_idë¥¼ ì°¾ì§€ ëª»í–ˆì„ ê²½ìš°
            if(playlist_id == ''):
                print(playlist_name)
                print("playlist_id error")
                continue
            
            input_list[1]=playlist_id
            
            video = youtube.playlistItems().list(
                playlistId = playlist_id,
                part = 'snippet',
                maxResults = 50 #ì—°ìŠµì´ë¯€ë¡œ íšŸìˆ˜ë¥¼ ì œí•œí•œë‹¤, í”„ë¡œì íŠ¸ ì§„í–‰ì‹œ maxê°’ì€ ì—†ë‹¤.
                )
            
            playlist_short_items=(video.execute())['items']
            while video:
                #print("ok")
                video = youtube.playlistItems().list_next(
                    video,
                    video.execute()
                )
                if not video:
                    break
                playlist_short_items+=(video.execute())['items']
            
            #ê°ê°ì˜ ì˜ìƒë³„ description or titleì„ í†µí•´ ìƒí˜¸ëª… ì¶”ì¶œ
            #ì—¬ê¸°ë¶€í„°ëŠ” ê° ì±„ë„ì˜ descriptionì´ ìƒì´í•˜ë¯€ë¡œ í•˜ë“œì½”ë”©ìœ¼ë¡œ ìƒí˜¸ëª…ì„ ì¶”ì¶œ
            each_extract(playlist_short_items,channel_name,input_list)
            
            #ê° í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ë§ˆë‹¤ playlists_dfì— ì ì¬
            playlist_input_list[0]=playlist_id
            playlist_input_list[1]=channel_id
            playlist_input_list[2]=playlist_name
            playlist_input_list[3]=True #videoì´ë¯€ë¡œ true
            playlists_df_input(playlist_input_list)
    
    #ì¬ìƒëª©ë¡ì´ ì—†ëŠ” ê²½ìš°(ì‡¼ì¸ )
    #####################################
    else: #(is_video_list[i] == 0)
        #ì±„ë„ì •ë³´ ì¶”ì¶œ
        channel_response = youtube.channels().list(
            part='contentDetails', 
            id=channel_id
            ).execute()

        uploads_playlist_id = channel_response['items'][0]['contentDetails']['relatedPlaylists']['uploads'] #ìµœì‹ ìˆœìœ¼ë¡œ ë‹¨ìˆœ ì¶”ì¶œ
        
        input_list[1]=uploads_playlist_id
        
        playlistitems = youtube.playlistItems().list(
            part='snippet', 
            playlistId=uploads_playlist_id, 
            maxResults=5 #ì—°ìŠµì´ë¯€ë¡œ íšŸìˆ˜ë¥¼ ì œí•œí•œë‹¤, í”„ë¡œì íŠ¸ ì§„í–‰ì‹œ maxê°’ì€ ì—†ë‹¤.
            )

        playlist_short_items=(playlistitems.execute())['items']
        while playlistitems:
            #print("ok")
            playlistitems = youtube.playlistItems().list_next(
                playlistitems,
                playlistitems.execute()
            )
            if not playlistitems:
                break
            playlist_short_items+=(playlistitems.execute())['items']
        
        #ê°ê°ì˜ ì˜ìƒë³„ description or titleì„ í†µí•´ ìƒí˜¸ëª… ì¶”ì¶œ
        #ì—¬ê¸°ë¶€í„°ëŠ” ê° ì±„ë„ì˜ descriptionì´ ìƒì´í•˜ë¯€ë¡œ í•˜ë“œì½”ë”©ìœ¼ë¡œ ìƒí˜¸ëª…ì„ ì¶”ì¶œ
        each_extract(playlist_short_items,channel_name,input_list)
        
        #ê° í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ë§ˆë‹¤ playlists_dfì— ì ì¬
        playlist_input_list[0]=uploads_playlist_id
        playlist_input_list[1]=channel_id
        playlist_input_list[2]='shorts'
        playlist_input_list[3]=True #videoì´ë¯€ë¡œ true
        playlists_df_input(playlist_input_list)
    
```
